```python
#creat dataset kendi datasetimizi yarattik
#class 1
x1=np.random.normal(25,5,1000)
y1=np.random.normal(25,5,1000)
#calss 2
x2=np.random.normal(55,5,1000)
y2=np.random.normal(60,5,1000)
#calss 3
x3=np.random.normal(55,5,1000)
y3=np.random.normal(15,5,1000)
print(x1,x2,x3,y1,y2,y3)
```

```python
x=np.concatenate((x1,x2,x3),axis=0)
y=np.concatenate((y1,y2,y3),axis=0)#concatereyt edildi birlestirildi ya da columns haline getirildi sutunlastirildi hepsi ayni sey
print(x1,x2,x3,y1,y2,y3)#goruntulemek icin
dictionary={"x":x,"y":y}
data=pd.DataFrame(dictionary)
```

# Different Data (Farkli data)

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.show()
```

# Similar Data (benzer data)

```python
plt.scatter(x1,y1,color="black")
plt.scatter(x2,y2,color="black")
plt.scatter(x3,y3,color="black")
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[] #kendi listemizi actik

for k in range(1,15): #for icinde 1 den 15 e kadar deneyecek
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_) #toplam formülünde ki yapi
```

```python
plt.plot(range(1,15),wcss)
plt.xlabel("number of k (cluster) values")
plt.ylabel("wcss")
plt.show()
```

```python
#k =3 icin denemeye yaptik

kmeans2=KMeans(n_clusters =3)
clusters=kmeans2.fit_predict(data)#kmeans2 degerlerini fit edip predictle ve clusters olarak ekle


data["label"]=clusters

plt.scatter(data.x[data.label == 0],data.y[data.label ==0],color ="red")
plt.scatter(data.x[data.label == 1],data.y[data.label ==1],color ="green")
plt.scatter(data.x[data.label == 2],data.y[data.label ==2],color ="blue")
plt.show()
```

# Cluster Centers

```python
#centerslarini bulduk(merkezlerini)
plt.scatter(data.x[data.label == 0],data.y[data.label ==0],color ="orange")
plt.scatter(data.x[data.label == 1],data.y[data.label ==1],color ="green")
plt.scatter(data.x[data.label == 2],data.y[data.label ==2],color ="yellow")
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color ="red")
plt.show()
```
